```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
days =['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
Money = [33.1, 32.5, 30.5, 28.3, 27.5, 26.5, 24.1]
Mood = [19.3, 24.9, 31.3, 33.0, 34.9, 34.3, 33.0]
```

```python
fig, ax = plt.subplots()
ax.set(xlabel='Day',ylabel='ratings',title='mood vs money comparison')
ax.plot(days, Mood)
ax.plot(days, Mood, "oy")
ax.plot(days, Money)
ax.plot(days, Money, "or")
plt.show()
```
